```python
avg_pnrg = []
avg_temp = []
avg_kinetic = []
```

```python
time, kinetic, pnrg, temp = np.loadtxt('energy.dat',dtype='float',unpack=True)
```

```python
kinetic = kinetic[-1]
pnrg = pnrg[-1]
temp = temp[-1]

avg_pnrg.append(pnrg)
avg_temp.append(temp)
avg_kinetic.append(kinetic)
```

```python
avg_pnrg = sorted(avg_pnrg)
avg_temp = sorted(avg_temp)
avg_kinetic = sorted(avg_kinetic)
```

```python
print(avg_pnrg)
print(avg_temp)
print(avg_kinetic)
```

```python
plt.figure(figsize=(10, 6))
plt.title('Potential Energy per atom - Temperature')
plt.xlabel('Temperature (K)')
plt.ylabel('Potential Energy per atom (eV)')
plt.scatter(avg_temp, np.array(avg_pnrg)/216, label='Potential Energy', color='blue')
plt.axvline(x=1665.5, color='k', linestyle='--')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Kinetic Energy per atom - Temperature')
plt.xlabel('Temperature (K)')
plt.ylabel('Kinetic Energy per atom (eV)')
plt.plot(avg_temp, np.array(avg_kinetic)/216, color='blue', linestyle='--')
plt.scatter(avg_temp, np.array(avg_kinetic)/216, label='Kinetic Energy', color='blue')
plt.legend()
plt.grid()
plt.show()
```
